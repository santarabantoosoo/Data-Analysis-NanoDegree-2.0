# 10. melt, pivot, stack, unstack

```python
import pandas as pd
```

```python
ad = pd.read_csv('data/airline_delay.csv')
ad
```

```python
ad.melt(id_vars='airline', value_vars=['ATL', 'DEN', 'DFW'])
```

```python
ad_idx = ad.set_index('airline')
ad_idx
```

```python
ad_idx.stack()
```

```python
ad_idx.stack().reset_index()
```

```python
ad.melt(id_vars='airline', var_name='airport', value_name='arrival delay')
```

```python
df = pd.read_csv('data/state_fruit.csv')
df
```

```python
df_idx = df.set_index(['state', 'fruit'])
df_idx
```

```python
df_idx.unstack()
```
